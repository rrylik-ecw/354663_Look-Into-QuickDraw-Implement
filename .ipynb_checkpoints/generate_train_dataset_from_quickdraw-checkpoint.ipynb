```python
!pip install google-cloud-storage
!pip install ndjson


# Set GCS credential
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "cricut-392320-97213bd8df64.json"


import json
from google.cloud import storage
import ndjson
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt


# Custom setting
default_size = 256.0
custom_size = (512, 512)
custom_back_color = (73, 109, 137)
custom_pen_color = (255, 255, 255)
custom_pen_width = 3


# Using the public quickdraw dataset
num_categories = 10
num_samples = 50
bucket_name = 'quickdraw_dataset'
folder_name = 'full/simplified'
download_path = "ndjson"
generated_data_folder = "generated_data"
json_file_name = "metadata.jsonl"
```

```python
# Get list of blobs
def list_blobs(bucket_name):
    """Lists all the blobs in the bucket."""
    storage_client = storage.Client()

    # Note: Client.list_blobs requires at least package version 1.17.0.
    blobs = storage_client.list_blobs(bucket_name)

    return blobs


# Get file list from subfolder in bucket
def list_files_in_subfolder(bucket_name, subfolder):
    """Lists all the files in a specific subfolder of the bucket."""
    storage_client = storage.Client()
    blobs = storage_client.list_blobs(bucket_name, prefix=subfolder)

    files = []
    for blob in blobs:
        files.append(blob.name)

    return files


# Download data from GCS bucket
def download_blob(bucket_name, source_blob_name, destination_file_name):
    """Downloads a blob from the bucket."""
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blob = bucket.blob(source_blob_name)

    blob.download_to_filename(destination_file_name)

    print(f"Blob {source_blob_name} downloaded to {destination_file_name}.")
    

blobs = list_files_in_subfolder(bucket_name, folder_name)

# Check if the directory exists
if not os.path.exists(download_path):
    # If it does not exist, create it
    os.makedirs(download_path)
    
# Download the file
ndjson_file_list = []
for i in range(num_categories):
    ndjson_file_path = download_path + '/' + blobs[i].split('/')[-1]
    ndjson_file_list.append(ndjson_file_path)
    download_blob(bucket_name, blobs[i], download_path + '/' + blobs[i].split('/')[-1])
```

```python
# Check if the directory exists
if not os.path.exists(generated_data_folder):
    # If it does not exist, create it
    os.makedirs(generated_data_folder)

with open(generated_data_folder + '/' + json_file_name, "w") as json_file:
    for no in range(len(ndjson_file_list)):
        destination_file_name = ndjson_file_list[no]
        # Open .ndjson file
        with open(destination_file_name) as f:
            data = ndjson.load(f)

        # Image ratio to default size
        ratio = custom_size[0] / default_size

        # For demonstration, we'll just take the first 5 drawings
        for i, drawing in enumerate(data[:num_samples]):
            # Each stroke is a continuous line
            img = Image.new('RGB', custom_size, color = custom_back_color) # You can adjust the size here
            d = ImageDraw.Draw(img)

            # Draw each stroke
            for stroke in drawing['drawing']:
                # Reformats the stroke into a list of points
                stroke[0] = [int(stroke[0][i]) * ratio for i in range(len(stroke[0]))]
                stroke[1] = [int(stroke[1][i]) * ratio for i in range(len(stroke[1]))]
                stroke_points = list(zip(stroke[0], stroke[1]))
                d.line(stroke_points, fill=custom_pen_color, width=custom_pen_width)

            # Save image with caption
            file_name = str((no + 1) * 1000000 + i)

            img.save(generated_data_folder + '/' + file_name + '.png')

            # Save caption to a text file
            with open(generated_data_folder + '/' + file_name + '.txt', 'w') as caption_file:
                caption_file.write(drawing["word"])

            # Prepare data for JSON
            json_data = {"id": int(file_name), "image": file_name + '.png', "text": "draw, " + ", ".join(drawing['word'].split("_"))}
            json.dump(json_data, json_file)
            json_file.write("\n") # Write each entry on a new line

            # Show the image along with the caption
            if i < 3:
                plt.figure()
                plt.title(drawing["word"])
                plt.imshow(img)
                plt.show()
```
